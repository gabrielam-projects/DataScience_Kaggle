```python
# Libraries and tools
import numpy as np # data 
import pandas as pd # data processing

from sklearn.preprocessing import MinMaxScaler # normalizer

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

# **1. Import dataset.**

```python
# Datasets in practice resources.
dataset_train = pd.read_csv('../input/titanic/train.csv')
dataset_test = pd.read_csv('../input/titanic/test.csv')
```

```python
### Delete columns in train and test datasets that doesn't contain impact information for learning or have too many missing data. ###

# Name -> Because it's not relevant.
# Ticket -> Cause maybe it's a random number.
# Cabin -> Only have 204/891 values (~22%). 
# PassengerId -> Because don't give us information and use it only in test set. 

deleted_colums_train = ["Name","Ticket","Cabin","PassengerId"]
dataset_train.drop(deleted_colums_train, inplace=True, axis = 'columns')
deleted_colums_test = ["Name","Ticket","Cabin"]
dataset_test.drop(deleted_colums_test, inplace=True, axis = 'columns')
```

```python
# Missing value for each column in train dataset.
print("Missing values in train columns:")
dataset_train.isnull().sum()
```

```python
# Missing value for each column in test dataset.
print("Missing values in test columns:")
dataset_test.isnull().sum()
```

```python
### Complete data in missing values. ###

## Colum "Age" in both datasets.
# Use average value to complete.
titanic_data_age = pd.concat([dataset_train["Age"], dataset_test["Age"]]) # Join all age values.
titanic_data_age = titanic_data_age.dropna() # Drop NaN values.
print("Average age in data to fill = ", round(titanic_data_age.mean())) # Calculate mean and round in an integer.

# Fill colum "Age"
dataset_train["Age"] = dataset_train["Age"].fillna(value = round(titanic_data_age.mean())) 
dataset_test["Age"] = dataset_test["Age"].fillna(value = round(titanic_data_age.mean()))



## Colum Embarked in train dataset.
# Counting how many times appear each value and fill with the max.
titanic_data_embarked = dataset_train["Embarked"].value_counts() 
max_value = titanic_data_embarked.index[0]
print("Most frequent value to fill: ", max_value)

# Fill colum "Embarked"
dataset_train["Embarked"] = dataset_train["Embarked"].fillna(value = max_value)



## Colum Fare in test dataset.
# Use average value to complete.
titanic_data_fare = pd.concat([dataset_train["Fare"], dataset_test["Fare"]]) # Join all fares values.
titanic_data_fare = titanic_data_fare.dropna() # Drop NaN values.
print("Average fare in data to fill = ", round(titanic_data_fare.mean(),2)) # Calculate mean and round in an integer.

# Fill colum "Fare" 
dataset_test["Fare"] = dataset_test["Fare"].fillna(value = round(titanic_data_fare.mean(),2))
```

# **Data normalization for training.**

```python
print("Training data info:")
dataset_train, dataset_train.info()
```

```python
# Convert Sex object "female"=0 and "male"=1.
dataset_train["Sex"].replace(["female","male"],[0,1],inplace=True)
dataset_test["Sex"].replace(["female","male"],[0,1],inplace=True)

# Convert Embarked object "S"=1, "C"=2 and "Q"=3.
dataset_train["Embarked"].replace(["S","C", "Q"],[1,2,3],inplace=True)
dataset_test["Embarked"].replace(["S","C", "Q"],[1,2,3],inplace=True)
```

```python
# Normalize some attributes in data between -1 and 1
scaler = MinMaxScaler()
dataset_train["Age"] = scaler.fit_transform(dataset_train["Age"].values.reshape(-1, 1))
dataset_train["Fare"] = scaler.fit_transform(dataset_train["Fare"].values.reshape(-1, 1))

dataset_test["Age"] = scaler.fit_transform(dataset_test["Age"].values.reshape(-1, 1))
dataset_test["Fare"] = scaler.fit_transform(dataset_test["Fare"].values.reshape(-1, 1))
```

# **3. Modelos de entrenamiento.**

```python
# Define X as input and y as output.

X = dataset_train.drop("Survived", axis=1)
y = dataset_train["Survived"]
print("Inputs: \n", X)
print("Output: \n", y)
```

```python
# Separate in train and test sets to training.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
## Logistic Regression

logreg = LogisticRegression(max_iter=250)
logreg.fit(X_train, y_train)
Y_pred = logreg.predict(X_test)
print('Score Logistic Regression:')
print(logreg.score(X_train, y_train))
```

```python
IDs = dataset_test["PassengerId"]
prediccion_logreg = logreg.predict(dataset_test.drop("PassengerId", axis=1))
out_logreg = pd.DataFrame({ "PassengerId" : IDs, "Survived": prediccion_logreg })
out_file = out_logreg.to_csv("out_logreg.csv")
print('Logistic Regression Prediction:')
out_logreg.head(15)
```
